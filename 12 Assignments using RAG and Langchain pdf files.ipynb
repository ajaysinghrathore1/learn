```python
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFaceHub
from langchain.vectorstores import Chroma

from langchain_anthropic import ChatAnthropic
```

```python
# Load and parse the PDFs
pdf_loader_1 = PyPDFLoader("nke-10k-2023.pdf")


# Load the documents as a list of text chunks
documents_1 = pdf_loader_1.load()


# Combine all documents
documents = documents_1 

```

```python
# Initialize a Hugging Face embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# # Create a FAISS vector store for fast retrieval of documents
# vector_store = FAISS.from_documents(documents, embedding_model)
# retriever = vector_store.as_retriever()
```

```python
CHROMA_PATH = "Chroma"
# embed the chunks as vectors and load them into the database
db_chroma = Chroma.from_documents(documents, embedding_model, persist_directory=CHROMA_PATH)
```

```python
# ----- Retrieval and Generation Process -----

# this is an example of a user question (query)
query = "What was Nike's revenue in 2023?"

# retrieve context - top 5 most relevant (closests) chunks to the query vector
# (by default Langchain is using cosine distance metric)
docs_chroma = db_chroma.similarity_search_with_score(query, k=5)

# generate an answer based on given user query and retrieved context information
context_text = "\n\n".join([doc.page_content for doc, _score in docs_chroma])
print(context_text)
```
