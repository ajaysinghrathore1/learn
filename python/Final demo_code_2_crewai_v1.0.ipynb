```python

# --- 3. Define the Agent ---
database_agent = Agent(
    role="Python Database Engineer",
    goal="Generate robust, reusable Python code for database connections.",
    backstory=(
        "You are an expert Python developer specializing in database connectivity. "
        "You write modular, well-documented, and error-handled code for any database system."
    ),
    # tools=[code_interpreter],
    allow_code_execution=True,
    llm=chat,  # Use the AzureChatOpenAI instance
    verbose=True,
)

# --- 4. Define the Task ---
db_connection_task = Task(
    description=(
        "Write a Python script that establishes a generic database connection. "
        "The script should:\n"
        "1. Accept database type, host, port, username, password, and database name as parameters.\n"
        "2. Use a modular approach (functions/classes).\n"
        "3. Include error handling and logging.\n"
        "4. Be well-documented with comments."
    ),
    expected_output="A Python script (as a string) that connects to any database using the provided parameters.",
    agent=database_agent,
)

# --- 5. Assemble the Crew ---
db_crew = Crew(
    agents=[database_agent],
    tasks=[db_connection_task],
    name="Database Connection Crew",
    description="A crew to generate modular, generic database connection scripts.",
    verbose=True,
)

# --- 6. Run the Crew ---
# crew_output = db_crew.kickoff(inputs={"create a prime number generator in python": "create a prime number generator in python"})
crew_output = db_crew.kickoff()

# Note: The input here is just a placeholder; the task description will guide the agent to generate the required script.
# If you want to see the output of the crew, you can print it   
# print(crew_output)
# or access the tasks_output attribute  
# to see the generated code.
# print(crew_output.tasks_output)   

# --- 7. Output the Result ---
print("Generated Database Connection Script:\n")
print(crew_output)
```
